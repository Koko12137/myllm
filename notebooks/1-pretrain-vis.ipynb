```python
import json

import matplotlib.pyplot as plt
```

# Load Train Logs

```python
log_file = 'output_03022108/checkpoint-34860/trainer_state.json'
with open(log_file, 'r') as f:
    log = json.load(f)['log_history']
log[0]
```

```python
losses = [x['loss'] for x in log]
steps = [x['step'] for x in log]
```

# Plot loss

```python
# Set up a figure and axis with a specific size
fig, ax = plt.subplots(figsize=(10, 6))

# Set title and labels
ax.set_title('Loss over Steps')
ax.set_xlabel('Steps')
ax.set_ylabel('Loss')
```

```python
# Plot the data
ax.plot(steps, losses)
```
